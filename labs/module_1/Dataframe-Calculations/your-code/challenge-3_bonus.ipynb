## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub problem 1
purchase_power = orders[['CustomerID', 'amount_spent']].groupby(['CustomerID']).agg(['sum', 'mean', 'min', 'max'])
purchase_power.reset_index(inplace=True)
```

```python
customer_ID = purchase_power[['CustomerID']]
```

```python
customer_ID = customer_ID.swaplevel(i=-2, j=-1, axis=1)
```

```python
purchase_power.drop(['CustomerID'], inplace=True, axis=1)
```

```python
final = pd.concat([customer_ID, purchase_power], axis=1)
final.columns = final.columns.droplevel()
```

```python
final
```

```python
# Sub problem 2
VIP_cutoff = np.quantile(final['sum'], 0.95)
Preferred_cutoff = np.quantile(final['sum'], 0.75)
filter_for_VIP_customers = final['sum']>VIP_cutoff
filter_for_preferred_customers = ((final['sum']>=Preferred_cutoff) & (final['sum']<=VIP_cutoff))
VIP_customers = final[filter_for_VIP_customers][['CustomerID', 'sum']]
Preferred_customers = final[filter_for_preferred_customers][['CustomerID', 'sum']]
```

```python
print(VIP_cutoff, '\n\n', VIP_customers)
```

```python
print(Preferred_cutoff, '\n\n', Preferred_customers)
```

## Q2: How to identify which country has the most VIP Customers?

```python
orders.head()
```

```python
customer_country = orders[['Country', 'CustomerID']].groupby(['CustomerID', 'Country']).count()
customer_country_list = customer_country.index.to_list()       
customer_country_dict = {tup[0]:tup[1] for tup in customer_country_list}
```

```python
VIP_customers['Country'] = VIP_customers['CustomerID'].map(customer_country_dict)
```

```python
VIP_customers[['Country', 'CustomerID']].groupby(['Country']).count().sort_values(by='CustomerID', ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
VIP_Preferred_combined = pd.concat([VIP_customers, Preferred_customers], ignore_index=True)
VIP_Preferred_combined
```

```python
VIP_Preferred_combined['Country'] = VIP_Preferred_combined['CustomerID'].map(customer_country_dict)
VIP_Preferred_combined
```
